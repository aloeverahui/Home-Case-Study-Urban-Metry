## 2.1 Income Range vs Population for Neighbourhood A, B, and C


### 2.1.1 Bar Chart 
Plot a Bar Chart to see the Income Range vs Population for Neighbourhood A, B, and C.

```python
# Set 'Neighbourhood' as the index for better plotting
df.set_index('Neighbourhood', inplace = True)

# Transpose the data for bar chart plotting
df_transposed = df.transpose()
df_transposed
```

```python
# Plot the bar chart
ax = df_transposed.plot(kind='bar', figsize=(12, 6))

# Set plot title and labels
plt.title('Income Range vs Population for Neighbourhood A, B, and C')
plt.xlabel('Income Range(RM)')
plt.ylabel('Population')
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

### 2.1.2 Heatmap
Plot a Heatmap to see the Income Range vs Population for Neighbourhood A, B, and C.

```python
# Assume there is average 1000 people in each neighbourhood to let our data value look quantifiable
df.iloc[:,:]= df.iloc[:,:]*1000

#Change the values from flaot to interger 
df.iloc[:,:]= df.iloc[:,:].astype(int)
```

```python
# Plot new Heatmap with new data for 1000 people averagely
plt.figure(figsize=(20,8))
heatmap = sns.heatmap(df, annot=True, fmt='.0f', cmap='inferno')

heatmap.set_xlabel('Income Range(RM)')
heatmap.set_ylabel('Neighbourhood')
heatmap.set_title('Income Range vs Population for Neighbourhood A, B, and C')

plt.show()
```

## 2.2 Comparison of Three Most Populated Income Range Between Neighbourhoods


### 2.2.1 Bar Chart 
Plot a Bar Chart to compare the Three Most Populated Income Range Within Neighbourhoods.

```python
# To find the three most populated income range between three neighbourhoods
column_sums = pd.Series(df.sum(axis=0))
sorted_column_sums = column_sums.sort_values(ascending=False)
print('The three most populated income range are:')
sorted_column_sums.head(3)
```

```python
# Define bar colors
income_range = ['3000-3999', '4000-4999', '5000-5999']
ax = df[income_range].plot(kind='bar', figsize=(10,6), color=['cyan', 'magenta', 'yellow'], edgecolor='black')

# Annotate each bar with its value (rounded to integer)
for p in ax.patches:
    ax.annotate(str(int(round(p.get_height(), 0))), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
    
plt.title('Comparison of Three Most Populated Income Range Between Neighbourhood A, B, and C')
plt.xlabel('Neighbourhood')
plt.ylabel('Population')
plt.xticks(rotation=0)

plt.tight_layout
plt.show()
```

## 2.3 Ratio of Each Neighbourhood in Top Populated Income Range

### 2.3.1 Waffle Chart
Plot a Waffle Chart to see the Ratio of Each Neighbourhood in Top Populated Income Range.

```python
# Install pywaffle
%pip install pywaffle
# Import module
from pywaffle import Waffle
# Create data for waffle chart
df_3000_3999 = df['3000-3999']
# Set up waffle chart figure
fig = plt.figure(
    FigureClass=Waffle,
    values=df_3000_3999,
    rows=4,
    columns = 13,
    icons='home',
    font_size=25,
    interval_ratio_x=2,
    interval_ratio_y=3,
    colors=['teal', 'gray', 'indigo'],
legend={
        'labels': ['A', 'B','C'], 
        'loc': 'upper left',
    'bbox_to_anchor': (0, -0.1)})

plt.show()
```
